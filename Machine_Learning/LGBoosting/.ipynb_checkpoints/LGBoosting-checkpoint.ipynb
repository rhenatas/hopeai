```python
import pandas as pd
```

```python
dataset = pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York' ]]
```

```python
independent
```

```python
dependent=dataset[['Profit']]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split   # import the library which is used for regression calculations
# Splitting and savinvg the training input and test values
# X_train - input train data
# X_test - input test data
# Y_train - output train data
# Y_test - output test data

X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from lightgbm import LGBMRegressor
import lightgbm as lgb


lgb_train = lgb.Dataset(X_train,Y_train )
lgb_eval = lgb.Dataset(X_test,Y_test, reference=lgb_train )
params = { 'boosting_type': 'gbdt',
          'objective': 'regression',
          'num_leaves': 20,
          'learning_rate': 0.1,
          'feature_fraction': 0.9   
}

gbm = lgb.train(params, lgb_train, num_boost_round=20, valid_sets=[lgb_train, lgb_eval], valid_names=['train','valid'], callbacks=[lgb.early_stopping(10)])
 
```

```python
# predictions

Y_pred = gbm.predict(X_test, gbm.best_iteration)

y_pred_binary = [1 if x > 0.5 else 0 for x in Y_pred]

Y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test, y_pred_binary)
accuracy
```
