```python
import pandas as pd
```

```python
dataset = pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York' ]]
```

```python
independent
```

```python
dependent=dataset[['Profit']]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split   # import the library which is used for regression calculations
# Splitting and savinvg the training input and test values
# X_train - input train data
# X_test - input test data
# Y_train - output train data
# Y_test - output test data

X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# Learning phase
from sklearn.tree import DecisionTreeRegressor # Importing Linear regression class and its procedures
regressor=DecisionTreeRegressor(criterion='poisson', splitter='random', max_features='log2') # Creating the object and assigning it to the regressor variable
regressor = regressor.fit(X_train,Y_train) # invoking the fit method to train the model / Create the model
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
# Passsing the test data and identifying the prediected value
Y_pred=regressor.predict(X_test)
```

```python
# Finding the R squeare value 
# importing the r square functionality from the library
from sklearn.metrics import r2_score

# passing the test data and the predicted data
r_score=r2_score(Y_test, Y_pred)
r_score
```
