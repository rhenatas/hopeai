```python
import pandas as pd  #Import the panda library which is used for file operations in python. It is referred as pd
```

```python
dataset = pd.read_csv("Salary_Data.csv")   # read the file and save the contents in the variable
```

```python
dataset 
```

```python
independent=dataset[["YearsExperience"]]  # read the file contents and save the input / independent values in the variable
dependent = dataset[["Salary"]]  # read the file contents and save the output / dependent values in the variable
```

```python
independent
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split   # import the library which is used for regression calculations
# Splitting and savinvg the training input and test values
# X_train - input train data
# X_test - input test data
# Y_train - output train data
# Y_test - output test data

X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)  
```

```python
# Learning phase
from sklearn.linear_model import LinearRegression # Importing Linear regression class and its procedures
regressor=LinearRegression() # Creating the object and assigning it to the regressor variable
regressor.fit(X_train,Y_train) # invoking the fit method to train the model / Create the model
```

```python
# Weight is the next predictor / slope
weight=regressor.coef_
weight
```

```python
# identifying the bais which is stored in the regressor variable
bais=regressor.intercept_
bais
```

```python
# Passsing the test data and identifying the prediected value
Y_pred=regressor.predict(X_test)
```

```python
# Finding the R squeare value 
# importing the r square functionality from the library
from sklearn.metrics import r2_score

# passing the test data and the predicted data
r_score=r2_score(Y_test, Y_pred)
r_score
```

```python
import pickle
filename="linear_final_model.sav"
```

```python
#save the model in a file
pickle.dump(regressor,open(filename,'wb'))
```

```python
#Open the model file in the read only mode
loaded_model=pickle.load(open("linear_final_model.sav",'rb'))
# save the predicted value in the variable
result=loaded_model.predict([[13]])
```
